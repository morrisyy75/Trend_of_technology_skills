```python
# Load the survey dataset
df = pd.read_csv('/kaggle/input/survey-m1/m1_survey_data.csv')
# Understand the dataset
df.head()
```

```python
# Number of duplicates in the dataset
df.duplicated().sum()
```

```python
# Remove duplicates
df_no_dup = df.drop_duplicates()
# Check 
print (df_no_dup.duplicated().sum())
df_no_dup.shape[0]
```

```python
# Check empty cells in all columns
print (df_no_dup.isnull().sum())
```

```python
#Imputing missing values
#Replace empty cells in the column by majority value
```

```python
# Finding majority in 'WorkLoc'
wlmax = df_no_dup['WorkLoc'].value_counts().idxmax()
print (wlmax)
```

```python
# Finding majority in 'Employment'
employmax = df_no_dup['Employment'].value_counts().idxmax()
print(employmax)
```

```python
# Finding majority in 'ConvertedComp'
ccmax = df_no_dup['ConvertedComp'].value_counts().idxmax()
print (ccmax)
```

```python
# Replace empty cells by majority values
# Rename the new df
df_no_dup['WorkLoc'] = df_no_dup['WorkLoc'].replace(np.nan,wlmax)
df_no_dup['Employment'] = df_no_dup['Employment'].replace(np.nan,employmax)
df_no_dup['ConvertedComp'] = df_no_dup['ConvertedComp'].replace(np.nan, ccmax)
# Check if any duplicants or empty cells in the dataframe
df_no_dup['WorkLoc'].isnull().sum()



```

```python
df_no_dup['CompFreq'].value_counts()
```

```python
df_no_dup['CompTotal'].value_counts()
```

```python
# Normalize the dataset to make data comparable.
df_no_dup['NormalizedSalary'] = df_no_dup.apply(lambda row: row['CompTotal'] * 52 if row['CompFreq'] == 'Weekly' else (
                                      row['CompTotal'] * 12 if row['CompFreq'] == 'Monthly' else row['CompTotal']),
                                  axis=1)
```
